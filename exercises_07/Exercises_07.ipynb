## Autocorrelation and blocks size ##

After the production of $M = 10^5$ instantaneous values of the systems's pontential energy and pressure in its various phases, the autocorrelation time is computed from this data in order to find the proper size for blocking averages.

First, the autocorrelation function $\chi(t)$ (for energy and pressure) is plotted for 300 steps and fitted with an exponential function to get an extimation of the correlation time $t_c$.
From the fits we obtain correlation times of around 10 steps in the solid and liquid phases, while in the gas phase the value is around 6 for internal energy and 1 for pressure.

Then, the statistical uncertainty is plotted as a function of the number of steps per block (logarithmic x scale). From these graphs we can see that a proper block size should be between 100 and 500 steps. In this range the uncertainty is maximal, while for bigger sizes starts to fluctuate excessively.

```python
#AUTOCORRELATION FUNCTION
def autocorrelation(var, steps):
    #ave = np.mean(var)
    sigma2 = np.var(var)
    t_max = len(var)
    
    chi = np.zeros(steps)
    
    for t in np.arange(0,steps):
        sum1 = 0.0
        sum2 = 0.0
        sum3 = 0.0
        for t1 in np.arange(0,t_max-t):
            sum1 += var[t1]*var[t+t1]
            sum2 += var[t1]
            sum3 += var[t+t1]    
        norm = t_max-t
        chi[t] = (sum1/norm-sum2/norm*sum3/norm)/sigma2
      
    return chi

#FIT FUNCTION
def f(x,k):  
    return np.exp(-x/k)

#BLOCKING AVERAGES FUNCTION
def blocking(var, L): 
    def error(AV,AV2,n):  # Function for statistical uncertainty estimation
        if n==0:
            return 0
        else:
            return np.sqrt((AV2[n] - AV[n]**2)/n)
        
    M = len(var) # total number of measurements
    N = int(M/L) # number of blocks
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = np.zeros(N)
    sum2_prog = np.zeros(N)
    err_prog = np.zeros(N)

    for i in range(N):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += var[k]
        ave[i] = sum/L       # r_i 
        av2[i] = (ave[i])**2 # (r_i)^2 
        
    for i in range(N):
        for j in range(i+1):
            sum_prog[i] += ave[j] # SUM_{j=0,i} r_j
            sum2_prog[i] += av2[j] # SUM_{j=0,i} (r_j)^2
        sum_prog[i]/=(i+1) # Cumulative average
        sum2_prog[i]/=(i+1) # Cumulative square average
        err_prog[i] = error(sum_prog,sum2_prog,i) # Statistical uncertainty
   
    return [sum_prog[N-1], err_prog[N-1]]

steps_max = 300
x = np.linspace(0,steps_max, steps_max)
```

### Solid ###

```python
ene_solid = np.loadtxt("./1autocor//solid/instant_epot.0", usecols=(0), unpack='true')
pres_solid = np.loadtxt("./1autocor/solid/instant_pres.0", usecols=(0), unpack='true')
```

```python
chi_ene_solid = autocorrelation(ene_solid, steps_max)
chi_pres_solid = autocorrelation(pres_solid, steps_max)
```

```python
popt_ene, pcov_ene= curve_fit(f, xdata=x, ydata=chi_ene_solid)
y_fit = f(x,popt_ene[0])
plt.plot(x,y_fit, 'r--',label=r'$t_c = {0:.2f} \pm {1:.2f}$'.format(popt_ene[0],np.sqrt(np.diagonal(pcov_ene))[0]))
plt.plot(chi_ene_solid)
plt.ylabel(r'$\chi(t)$')
plt.xlabel('t')
#plt.yscale('log')
plt.title('Solid: energy autocorrelation')
plt.legend()
plt.show()

popt_pres, pcov_pres= curve_fit(f, xdata=x, ydata=chi_pres_solid)
y_fit = f(x,popt_pres[0])
plt.plot(x,y_fit, 'r--', label=r'$t_c = {0:.2f} \pm {1:.2f}$'.format(popt_pres[0],np.sqrt(np.diagonal(pcov_pres))[0]))
plt.plot(chi_pres_solid)
plt.ylabel(r'$\chi(t)$')
plt.xlabel('t')
#plt.yscale('log')
plt.title('Solid: pressure autocorrelation')
plt.legend()
plt.show()
```

```python
#l_values = [10, 16, 20, 25, 32, 40, 50, 80, 100, 125, 160, 200, 250, 400, 500, 625, 800, 1000, 1250, 2000, 2500, 3125, 4000, 5000]
#sigma_ene_solid = np.zeros(len(l_values))

sigma_ene_solid = np.zeros(500)
l_values = np.zeros(500)

k=0
l = 10
while l <= 5000:
    l_values[k] = l
    temp = blocking(ene_solid, l)
    sigma_ene_solid[k] = temp[1]
    k += 1
    l += 10
    
sigma_pres_solid = np.zeros(500)
l_values = np.zeros(500)

k=0
l = 10
while l <= 5000:
    l_values[k] = l
    temp = blocking(pres_solid, l)
    sigma_pres_solid[k] = temp[1]
    k += 1
    l += 10
```

```python
plt.plot(l_values,sigma_ene_solid)
plt.ylabel(r'$\sigma$')
plt.xlabel('#Steps per block')
plt.xscale("log")
plt.title('Solid: energy uncertainty as a function of #steps per block')
plt.show()

plt.plot(l_values,sigma_pres_solid)
plt.ylabel(r'$\sigma$')
plt.xlabel('#Steps per block')
plt.xscale("log")
plt.title('Solid: pressure uncertainty as a function of #steps per block')
plt.show()
```

# Exercise 07.4 #

To compare Monte Carlo NVT and Molecular Dynamics NVE simulations of the same Lennard-Jones system, the number of steps and block are choosen to have the same number of "measurments" in both cases:
- Monte Carlo NVT
     - 200 steps per block
     - 50 blocks
- Molecular Dynamics NVE
    - 2000 steps per block that results in 200 measurments per block
    - 50 blocks

The results obtained with the two different approaches are comparable in the various phases. The radial distribution function is always almost identical, while the potential energy and pressure plot do not always overlap due the difficulty to equilibrate the MD simulation at an exact given temperature (the system is simulated in the microcanonical ensemble, so at fixed energy). Anyway the values of the observables that we get from the two approaches are very close.
