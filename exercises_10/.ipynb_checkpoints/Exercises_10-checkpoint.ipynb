## Simulated annealing implementation ##

A combinatorial optimization problem like TSP can be mapped into an annealing problem. 
We already have a cost function - i.e the path length $L^2(x)$ - that will become the energy of a configuration of the system to anneal.
In this way we can sample the Boltzamnn distribution for the system that, in the limit $\beta \to \infty $, will represent the ground state and thus the minimal value of $L^2(x)$.

To sample Boltzmann distribution we are going to use Metropolis algorithm.
A random move is generated using the mutation operators of the genetic algorithm (pair permutation and inversion, while total shift does not change the energy). At every step, randomly, only a type of mutation is executed.
The new mutated configuration will then replace the old with probability
$$ p=  \min (1, \exp (- \beta \Delta E)).$$

At this point to actually anneal the system we have to choose an annealing schedule: the temperatures that the system is going to visit and the time duration for each one.

I implemented a constant geometric annealing of the type
$$ \beta_{i+1} = \frac{\beta_i}{r},$$
where the initial $\beta_0 = 1 $, the cooling rate $r = 0.95$ and the system is cooled every 1000 steps.

The Metropolis sampling is embeded in the method `Move()` of the class `Annealing`, while the annealing is called every `ncool` steps with the method `SetBeta(new_beta)`.

TSP is realized in the same way of exercise 9 but, this time, a single object of type `Route` is created and not a collection of them. Optimization is carried out mutating this single instance and accepting or refusing these moves as prescribed by Metropolis algorithm.

### On a circle ###

```python
step_c, beta_c, min_c, = np.loadtxt("./results/sa_circle_30/output.len.0", usecols=(0,1,2), unpack='true')

fig, ax1 = plt.subplots()
ax1.plot(step_c, min_c, label='Length')
ax1.set_xlabel('#Step')
ax1.set_ylabel('$L^2$')

ax2 = ax1.twinx()
ax2.set_ylabel(r'$\beta$')
ax2.plot(step_c, 1/beta_c, label=r'Temperature', color='r')
fig.legend(loc=1, bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)

plt.title("TSP on a circle: simulated annealing")
plt.show()
```

```python
x_c, y_c = np.loadtxt("./results/sa_circle_30/frames/path_100000.xy", usecols=(1,2), unpack='true')

distance_c = 0
end_c = len(x_c) -1
for i in range(end_c):
    distance_c +=(x_c[i]-x_c[i+1])**2 + (y_c[i]-y_c[i+1])**2
distance_c += (x_c[end_c]-x_c[0])**2 + (y_c[end_c]-y_c[0])**2

plt.plot(x_c,y_c, '-r', label='$L^2 = {0:.3f}$'.format(distance_c))
plt.plot(x_c,y_c, 'ok')
plt.axis('scaled')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title("TSP on a circle: best path after $10^5$ SA steps")
plt.show()
```

## Parallel simulated annealing ##

As a first step, using MPI libraries, I parallelized the SA code. 

The result is a set of indipendent simulations that do not comunicate with each other and are different just for the random generator seed. At every step the legths of the configurations of the various processes are gathered in the root process to get the global minimal lenght found in the run: no exchange occur among the parallel processes.

Both the final configurations of each process and the best configuration found during the annealing are showed.

```python
step1_psa, beta1_psa, min1_psa, = np.loadtxt("./results/psa_square_30/output.len.1", usecols=(0,1,2), unpack='true')
step2_psa, beta2_psa, min2_psa, = np.loadtxt("./results/psa_square_30/output.len.2", usecols=(0,1,2), unpack='true')
step3_psa, beta3_psa, min3_psa, = np.loadtxt("./results/psa_square_30/output.len.3", usecols=(0,1,2), unpack='true')
step4_psa, beta4_psa, min4_psa, = np.loadtxt("./results/psa_square_30/output.len.4", usecols=(0,1,2), unpack='true')

step_glb_psa, min_glb_psa = np.loadtxt("./results/psa_square_30/output.len.0", usecols=(0,2), unpack='true')
```

```python
fig, ax1 = plt.subplots()
ax1.plot(step1_psa,min1_psa,label =r'P1 Length')
ax1.plot(step2_psa,min2_psa,label =r'P2 Length' )
ax1.plot(step3_psa,min3_psa,label =r'P3 Length')
ax1.plot(step4_psa,min4_psa,label =r'P4 Length')
ax1.plot(step_glb_psa,min_glb_psa, 'kx',label ='Best path found')

ax1.set_xlabel('#Step')
ax1.set_ylabel('$L^2$')

#ax1.legend(loc='upper right')
ax2 = ax1.twinx()
ax2.set_ylabel('$T$')
ax2.plot(step1_psa, 1/beta2_psa, label='Temperature', color='y')
#ax2.legend(loc='center right')
fig.legend(loc=1, bbox_to_anchor=(1,1), bbox_transform=ax1.transAxes)
plt.title("TSP in a square: parallel simulated annealing")
plt.show()
```

```python
x_s1, y_s1 = np.loadtxt("./results/psa_square_30/frames/path-1_100000.xy", usecols=(1,2), unpack='true')
x_s2, y_s2 = np.loadtxt("./results/psa_square_30/frames/path-2_100000.xy", usecols=(1,2), unpack='true')
x_s3, y_s3 = np.loadtxt("./results/psa_square_30/frames/path-3_100000.xy", usecols=(1,2), unpack='true')
x_s4, y_s4 = np.loadtxt("./results/psa_square_30/frames/path-4_100000.xy", usecols=(1,2), unpack='true')

def path_length(x,y):
    distance = 0
    end = len(x) -1
    for i in range(end):
        distance += (x[i]-x[i+1])**2 + (y[i]-y[i+1])**2
    distance += (x[end]-x[0])**2 + (y[end]-y[0])**2
    return distance

plt.plot(x_s1,y_s1, '-', linewidth=.8, label='P1: $L^2 = {0:.3f}$'.format(path_length(x_s1, y_s1)))
plt.plot(x_s2,y_s2, '-', linewidth=.8, label='P2: $L^2 = {0:.3f}$'.format(path_length(x_s2, y_s2)))
plt.plot(x_s3,y_s3, '-', linewidth=.8, label='P3: $L^2 = {0:.3f}$'.format(path_length(x_s3, y_s3)))
plt.plot(x_s4,y_s4, '-', linewidth=.8, label='P4: $L^2 = {0:.3f}$'.format(path_length(x_s4, y_s4)))

plt.plot(x_s1,y_s1, 'ok')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.title("TSP in a square: final paths after $10^5$ steps (4 indipendent annealing)")
plt.show()

x_best_psa, y_best_psa = np.loadtxt("./results/psa_square_30/frames/path-1_99000.xy", usecols=(1,2), unpack='true')
plt.plot(x_best_psa, y_best_psa, '-', label='$L^2 = {0:.3f}$'.format(path_length(x_best_psa, y_best_psa)))
plt.plot(x_best_psa, y_best_psa, 'ok')
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.title("TSP in a square: best path found after $10^5$ steps (4 indipendent annealing)")
plt.show()
```

## Parallel tempering ##

Then I implemented parallel tempering.

Now each process - a replica of the system - runs a simulation at a fixed temperature.
Every _nswap_ (= 100) steps two random adiacent replicas try to swap their configuration.
Choosing the set of temperatures properly (containing both high and low temperature and with a small enough spacing), an acceptance rate for the swap move around 0.5 can be obatained.

At every step both the configuration and the lengts of the replicas are gathered in the root process, which performs the swap attempt and sends back the configurations (swapped or not) to every replica.

The swap probability _p_, choosen to satisfy the detailed balance condition, is:
$$ p = \min (1, \exp (\Delta \beta \Delta E)), $$
where $\Delta \beta = \beta_i -\beta_{i-1}$ and $\Delta E=L^2(x_i)-L^2(x_{i-1})$.

The results show that with just $10^4$ steps - and thus a really faster execution time - for 8 replicas we can get an optimal path better than the one obtained with $10^5$ annealing steps and the one obtained from 4 parallel simulated annealing.
So parallel tempering, with a proper choice of the set of temperatures, seems to be the most efficient of the optimization algorithms implemented to solve TSP.

In the following charts are plotted:
- the lenghts (energies) of each replica during the simulation, with the positions of the accepted swaps among them;
- the probability distributions of the sampled replicas, which - as expected - partially overlap;
- the global best path found.
