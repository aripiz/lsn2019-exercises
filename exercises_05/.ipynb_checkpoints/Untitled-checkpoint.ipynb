```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

steps, orbital, delta, acc  = np.loadtxt("./test/acceptance_300.out", usecols=(0), unpack='true')

X,Y,Z = np.loadtxt("./test/coordinates_300.xyz", usecols=(0,1,2), unpack='true')




fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(90, 30)
plt.show()

R, error_R = np.loadtxt("./test/r_300.out", usecols=(0,1), unpack='true')
N = len(R)


plt.errorbar(range(N),R,yerr=error_R, label=r'$\langle r \rangle$')

plt.xlabel('#Steps')
plt.ylabel(r'$\frac{r}{a_0}$')
plt.legend(loc='upper right')
plt.grid(True)
plt.show()

print("Results:")
print("")
print("Step lenght = {0:1.2f}".format(delta))
print("<r/a0> = {0:2.3f} +/- {1:.3f}".format(R[N-1], error_R[N-1])) 
print("Acceptance rate = {0:2.2f}".format(acc))

```
