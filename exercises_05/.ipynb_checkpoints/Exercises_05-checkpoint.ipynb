# Exercise 05.1 #

## Introduction ##
Monte Carlo Metropolis estimation of $\langle r \rangle$ for 1s and 2p Hydrogen atom orbitals.

The central point is to sample $|\Psi_{nlm}(x,y,z)|^2 = p_{nlm}(x,y,z)$. 
A random move of lenght $\delta$ from $(x,y,z)$ to $(x',y',z')$ is implemented through:
1. a uniform transition probability (i.e. randomly choosing a point on the sphere of radius $\delta$ and centre $(x,y,z)$);
2. a normal transition probability (i.e. doing a step of lenght $\mathcal{N}(0,\delta/2)$ for each coordinate: $(x',y',z')=(x+\mathcal{N}(0,\delta/2), y+\mathcal{N}(0,\delta/2), z+\mathcal{N}(0,\delta/2))$).

The lenght $\delta$ of the step has been adjusted in order to obtain an acceptance rate $\alpha$ around 50% in each case, where, in the Metropolis algorithm, $$\alpha = \min \left[1,\frac{p(\mathbf{x'})}{p(\mathbf{x})}\right]$$


Working in Bohr radius units ($a_0=1$),
- $p_{1s}(x,y,z) = p_{1s}(r) = \frac{1}{\pi}e^{-2r}$
- $p_{2p}(x,y,z) = p_{2p}(r,\theta) = \frac{r^2}{32\pi}e^{-r}\cos^{2}\theta$

with $r^2 = x^2 + y^2 +z ^2 $ and $\theta = \tan^{-1}\frac{y}{x}$.

Sampling is carried out for $100$ blocks consisting of $10^6$ Metropolis steps.

```python
#Acceptance rates and step values importing
steps, delta1s, acc1s, delta2p, acc2p, delta1sN,acc1sN,r1sN, delta2pN,acc2pN,r2pN, delta320, acc320, delta321, acc321  = np.loadtxt("./ex_1/acceptance.out", usecols=(0), unpack='true')
```

## Uniform transition probability ##

### 1s orbital ###

```python
X,Y,Z = np.loadtxt("./ex_1/coordinates_1s.xyz", usecols=(0,1,2), unpack='true')
fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.title('Hydrogen 1s orbital')
plt.show()

R, error_R = np.loadtxt("./ex_1/r_1s.out", usecols=(0,1), unpack='true')
N = len(R)

plt.errorbar(range(N),R,yerr=error_R, label=r'$\langle r \rangle_{1s}$')

plt.xlabel('#Blocks')
plt.ylabel(r'$\frac{r}{a_0}$')
plt.legend(loc='upper right')
plt.grid(True)
plt.title('Hydrogen 1s orbital: average radius')
plt.show()

print("Average final values:")
print("Step lenght = {0:1.2f}".format(delta1s))
print("<r/a0> = {0:2.3f} +/- {1:.3f}".format(R[N-1], error_R[N-1])) 
print("Acceptance rate = {0:2.2f}".format(acc1s))
```
