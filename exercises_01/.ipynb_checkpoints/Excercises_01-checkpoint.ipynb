# Exercise 01.2 #

Central Limit Theorem check for different probability distribution.

$M=10^4$ realizations of $S_N = \frac{1}{N}\sum_{i=1}^N x_i$, as $N$ increases the CLT emerges for all the sampled distribution.

## Standard dice ##

```python
n_bins = 100

x1 = np.loadtxt("./ex_2/std_dice_N-1.out", usecols=(1), unpack='true')
x2 = np.loadtxt("./ex_2/std_dice_N-2.out", usecols=(1), unpack='true')
x10 = np.loadtxt("./ex_2/std_dice_N-10.out", usecols=(1), unpack='true')
x100 = np.loadtxt("./ex_2/std_dice_N-100.out", usecols=(1),unpack='true')


plt.hist(x1, n_bins, range=(1,6), label='N=1')
plt.hist(x2, n_bins, range=(1,6), label='N=2')
plt.hist(x10, n_bins, range=(1,6),  label='N=10')
plt.hist(x100, n_bins, range=(1,6), label='N=100')

plt.xlabel('$S_N$')
plt.ylabel('Frequency')
plt.title('Standard dice (random integer between 1 and 6)')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()
```

## Exponential dice ##

```python
x1 = np.loadtxt("./ex_2/exp_dice_N-1.out", usecols=(1), unpack='true')
x2 = np.loadtxt("./ex_2/exp_dice_N-2.out", usecols=(1),  unpack='true')
x10 = np.loadtxt("./ex_2/exp_dice_N-10.out", usecols=(1), unpack='true')
x100 = np.loadtxt("./ex_2/exp_dice_N-100.out", usecols=(1), unpack='true')

n_bins = 100

plt.hist(x1, n_bins, range=(0,4), label='N=1')
plt.hist(x2, n_bins, range=(0,4), label='N=2')
plt.hist(x10, n_bins, range=(0,4), label='N=10')
plt.hist(x100, n_bins, range=(0,4),  label='N=100')


plt.xlabel('$S_N$')
plt.ylabel('Frequency')
plt.title('Exponential distribution')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()

#FIT
from scipy.optimize import curve_fit


def f(x,a,b,c):  # defining the fitting function
    return a * np.exp(-b*(x-c)**2)

n, bins, patches = plt.hist(x100, n_bins, range=(0,2))
x_spacing = np.linspace(0, 2, n_bins)
popt, pcov= curve_fit(f, xdata=x_spacing, ydata=n)

y_fit = f(x_spacing,popt[0],popt[1],popt[2])


plt.plot(x_spacing,y_fit, 'r-')
plt.xlabel('$S_{100}$')
plt.ylabel('Frequency')
plt.title('Fit for N=100')
plt.grid(True)                   
plt.show()

print("optimized parameters [height, 1/2*sigma^2, mean] =")
print(popt[0], popt[1],popt[2])
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(pcov)))
```

## Lorentzian dice ##

```python
x1 = np.loadtxt("./ex_2/lor_dice_N-1.out", usecols=(1), unpack='true')
x2 = np.loadtxt("./ex_2/lor_dice_N-2.out", usecols=(1), unpack='true')
x10 = np.loadtxt("./ex_2/lor_dice_N-10.out", usecols=(1), unpack='true')
x100 = np.loadtxt("./ex_2/lor_dice_N-100.out", usecols=(1), unpack='true')

n_bins = 100


plt.hist(x1, n_bins, range=(-10,10), label='N=1')
plt.hist(x2, n_bins, range=(-10,10), label='N=2')
plt.hist(x10, n_bins, range=(-10,10),  label='N=10')
plt.hist(x100, n_bins, range=(-10,10), label='N=100')


plt.xlabel('$S_N$')
plt.ylabel('Probability')
plt.title('Lorentzian')
plt.grid(True)
plt.legend(loc='upper right')
plt.show()


#FIT
from scipy.optimize import curve_fit

def f(x,mean,gamma,height):  # defining the fitting function
    return height*(gamma/np.pi)/((x-mean)**2+gamma**2)


n, bins, patches = plt.hist(x100, n_bins, range=(-10,10))

x = np.linspace(-10, 10, n_bins)

popt, pcov= curve_fit(f, xdata=x, ydata=n)

y_fit = f(x,popt[0], popt[1], popt[2])


plt.plot(x,y_fit, 'r-')
plt.xlabel('$S_{100}$')
plt.ylabel('Frequency')
plt.title('Fit for N=100')
plt.grid(True)                   
plt.show()

print("optimized parameters [mean, gamma, height] =")
print(popt)
print("parameters uncertainty =")
print(np.sqrt(np.diagonal(pcov)))
```

# Exercise 01.3 #

Buffon's needle experiment to extimate $\langle \pi \rangle$. 

The used ratio between line spacing and needle lenght is $\frac{l}{d}= 0.25$.
To avoid using $\pi$ in the sampling of a random angle in the range $[0,\pi]$, the following code was implemented:

`  do{
      x_angle = rnd.Rannyu();
      y_angle = rnd.Rannyu();
    } while(x_angle*x_angle + y_angle*y_angle > 1);
    phi[i] = atan(y_angle/x_angle);
`

In this way the angle is sampled through an accept-reject technique that does not invole the explict use of the value of $\pi$.

$M=100000$ throws, divided in $N=100$ blocks.

```python
x, pi, err_prog = np.loadtxt("./ex_3/ex_3.out", usecols=(0,1,2), delimiter=';', unpack='true')
plt.errorbar(x,pi,yerr=err_prog)

plt.xlabel('#Blocks')
plt.ylabel(r'$\langle \pi \rangle$')
plt.grid(True)
plt.title("Buffon's needle experiment: extimation of $\pi$")
plt.show()
```
