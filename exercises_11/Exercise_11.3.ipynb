```python
import numpy as np

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
plt.rcParams["figure.figsize"] = (9.5,6)
```

```python
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects
```

## Input data creation ##

Function to fit:
$$ f(x,y) = \sin (x^2 +y^2) $$
$$ x \in [-\frac{3}{2},+\frac{3}{2}] $$

Number of input data:
- $N_{train} = 4000 $
- $N_{validation} = 400 $

Amount of normal noise in the data:
- $\sigma = 0.2 $


```python
#function to fit
def f(x,y):
    return np.sin(x**2+y**2)

lowb = -3/2
uppb = 3/2
```

```python
#number of input data
ntrain = 4000
nvalid = 400

# noise standard deviation
sigma = 0.2
```

```python
# generate training inputs and validation inputs
seed=0
np.random.seed(seed)
tf.set_random_seed(seed)

x_train = np.zeros((ntrain,2)) # input is composed of 2 coordinates!
x_train[:,0] = np.random.uniform(lowb, uppb, ntrain) 
x_train[:,1] = np.random.uniform(lowb, uppb, ntrain)

x_valid = np.zeros((nvalid,2)) # input is composed of 2 coordinates!
x_valid[:,0] = np.random.uniform(lowb, uppb, nvalid) 
x_valid[:,1] = np.random.uniform(lowb, uppb, nvalid)

# target function
y_target = f(x_valid[:,0],x_valid[:,1]) 

# actual measures from which we want to guess regression parameters
y_train = np.random.normal(f(x_train[:,0], x_train[:,1]), sigma) 
y_valid = np.random.normal(f(x_valid[:,0], x_valid[:,1]), sigma)
```

```python
# plot training and target dataset
X = np.arange(lowb, uppb, 0.1)
Y = np.arange(lowb, uppb, 0.1)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

fig = plt.figure()
ax = Axes3D(fig)

ax.scatter3D(x_valid[:,0], x_valid[:,1], y_valid, label='Validation')
ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,linewidth=0, antialiased=False)
#ax.plot_trisurf(x_valid[:,0], x_valid[:,1], y_target, color ='r')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.legend()
plt.show()
```

 ## Model construction and training ##

```python
nepoch = 100

#create model
model = tf.keras.Sequential()
model.add(Dense(8, input_shape=(2,), activation='relu'))

model.add(Dense(32, activation='tanh'))
model.add(Dense(1640, activation='relu'))

model.add(Dense(1))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

model.summary()
```

```python
# fit the model using training dataset
# report training progress against validation data

history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=nepoch,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

## Training results and prediction ##

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

# evaluate model with the va curve
score_valid = model.evaluate(x_valid, y_valid, batch_size=32, verbose=0)
score_target = model.evaluate(x_valid, y_target, batch_size=32, verbose=0)
print()
print('Model performance:')
print('Test loss with validation data = {0:.3f}' .format(score_valid[0]))
print('Test loss with target data = {0:.3f}' .format(score_target[0]))
```

```python
# plot predicted and target dataset inside training range
X = np.arange(lowb, uppb, 0.1)
Y = np.arange(lowb, uppb, 0.1)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

x_predicted = np.zeros((1000,2))
x_predicted[:,0] = np.random.uniform(lowb, uppb, 1000) 
x_predicted[:,1] = np.random.uniform(lowb, uppb, 1000)
y_predicted = model.predict(x_predicted)

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,linewidth=0, antialiased=False, alpha=0.5)
ax.scatter3D(x_predicted[:,0], x_predicted[:,1], y_predicted, label='Predicted')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.legend()
plt.title('Prediction inside training interval')
plt.show()

# plot predicted and target dataset outisde training range
X = np.arange(-3, 3, 0.1)
Y = np.arange(-3, 3, 0.1)
X, Y = np.meshgrid(X, Y)
Z = f(X,Y)

x_predicted = np.zeros((1000,2))
x_predicted[:,0] = np.random.uniform(-3, 3, 1000) 
x_predicted[:,1] = np.random.uniform(-3, 3, 1000)
y_predicted = model.predict(x_predicted)

fig = plt.figure()
ax = Axes3D(fig)

ax.plot_surface(X, Y, Z,cmap=cm.coolwarm,linewidth=0, antialiased=False, alpha=0.5)
ax.scatter3D(x_predicted[:,0], x_predicted[:,1], y_predicted, label='Predicted')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.legend()
plt.title('Prediction outside training interval')
plt.show()
```

The model - composed of 3 layers of size 8,32,16 - was trained for 100 epochs using SGD as optimization function and sequence of _relu, tanh, relu_ as activation functions.

The results show - as expected - that inside the training interval the model predictive capability is good, while outside is very poor.
