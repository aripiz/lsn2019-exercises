# Exercises 04.1- 2- 3 #

## Implementation of the Molecular Dynamics code ##

The ability to restart the simulation from previous spatial configuration was added.
So, the code has an option to start from a previous final configuration (`config.final` and `old.final`) and to rescale it or not in order to match the desired temperature.
To help carring out the initial equilibration procedure, 3 bash script are present in the folder:
 - `reset.sh` deletes every output files and copies `config.fcc` to `config.0`;
 - `restart.sh` copies `config.final` to `config.0` and `old.final` to `old.0`;
 - `clean.sh` deletes output files to start a clean simulation after equilibration.

Then blocking averages were added to the code for the following observables: potential energy, kinetic energy, total energy, temperature, pressure.

Update: tail correction for potential energy and pressure and radial distribution function calculation were implemeted.

## Equilibration at desidered temperature from FCC configuration ##

In the following example, regarding the system in the solid phase, the simulation was run:
- for 1000 steps from FCC configuration rescaling the old configuration to match a higher temperature (1.2);
- for 1000 steps from previous final configuration rescaling the old configuration to match desired temperature;
- for 2000 steps from previous final configuration without initial rescaling.

A similar procedure was followed to equilibrate the simulations of exercise 4.

```python
instant_temp= np.loadtxt("./results/equilibration/instant.temp.0", usecols=(0), unpack='true')
#instant_temp= np.loadtxt("./md_nve/output/instant.temp.0", usecols=(0), unpack='true')
plt.axhline(0.8, color='r', label='Target temperature')

plt.plot(instant_temp, label='Actual temperature')
plt.ylabel('Temperature*')
plt.xlabel('#Step/10')
plt.title('Equilibration example')

plt.legend()
plt.show()
```

# Exercises 04.4 #

MD simulation of a Lennard-Jones system of 108 particles in different phases for:
- 0.0005 as time step length (in reduced units)
- 2000 steps per block that results in 200 measurments per block
- 50 blocks

Phases:
 - Solid: $\rho^\star = 1.1$, $T^\star = 0.8$, $r_c = 2.2$ 
 - Liquid: $\rho^\star = 0.8$, $T^\star = 1.1$, $r_c = 2.5$
 - Gas: $\rho^\star = 0.05$, $T^\star = 1.2$, $r_c = 5.0$
 
Results are plotted for Argon and Krypton using SI units (see cell below for reference).

```python
#CONSTANTS
kB = 1.3806e-23
amu = 1.66054e-27

delta = 0.0005 #timestep
nstep = 2000 # number of steps per block
nblock = 50 # number of blocks

#ARGON 
sigma_Ar = 0.34e-9
m_Ar = 39.948*amu
eps_Ar = 120 *kB
delta_Ar = delta*np.sqrt(m_Ar*sigma_Ar**2/eps_Ar) 
time_Ar = np.zeros(nblock)
for i in range(len(time_Ar)):
    time_Ar[i] = i * delta_Ar*(nstep)

#KRYPTON
sigma_Kr = 0.364e-9
m_Kr = 83.798*amu
eps_Kr = 164 *kB
delta_Kr = delta*np.sqrt(m_Kr*sigma_Kr**2/eps_Kr)
time_Kr = np.zeros(nblock)
for i in range(len(time_Kr)):
    time_Kr[i] = i * delta_Kr*(nstep)
```

## Solid phase ##

```python
#SIMULATION PARAMETERS 
T_solid, npart, rho_solid, rcut_solid = np.genfromtxt("./results/solid_md/input.dat", usecols=(0), unpack='true', max_rows=4)

#IMPORTING SIMULATION RESULTS
block_solid, epot_solid, err_epot_solid = np.loadtxt("./results/solid_md/output.epot.0", usecols=(0,2,3), unpack='true')
block_solid, pres_solid, err_pres_solid = np.loadtxt("./results/solid_md/output.pres.0", usecols=(0,2,3), unpack='true')
block_solid, ekin_solid, err_ekin_solid = np.loadtxt("./results/solid_md/output.ekin.0", usecols=(0,2,3), unpack='true')
block_solid, temp_solid, err_temp_solid = np.loadtxt("./results/solid_md/output.temp.0", usecols=(0,2,3), unpack='true')
block_solid, etot_solid, err_etot_solid = np.loadtxt("./results/solid_md/output.etot.0", usecols=(0,2,3), unpack='true')

print("SIMULATION CONDITIONS (LJ reduced units):")
print("rho* = {0:.1f}".format(rho_solid))
print("T* = {0:.1f}".format(T_solid))
print("Rcut* = {0:.1f}".format(rcut_solid))
```

```python
solid_eq_temp= np.loadtxt("./results/eq/instant.temp.0", usecols=(0), unpack='true')
#solid_eq_etot= np.loadtxt("./md_nve/output/instant.temp.0", usecols=(0), unpack='true')

plt.axhline(T_solid, color='r', label='Target temperature')

plt.plot(solid_eq_temp, label='Actual temperature')
plt.ylabel('Temperature*')
plt.xlabel('#Step/10')
plt.title('Solid phase equilibration')

plt.legend()
plt.show()
```

```python
plt.errorbar(block_solid, etot_solid, yerr=err_etot_solid)
plt.errorbar(block_solid,epot_solid,yerr=err_epot_solid)
plt.errorbar(block_solid,ekin_solid, yerr=err_ekin_solid)
plt.xlabel('#Block')
plt.ylabel('Total Energy*')
plt.title('Conservation of energy')
plt.show()
```

### Argon ###

```python
print("ARGON:")
print("m = {0:.3e} kg".format(m_Ar))
print("rho = {0:.3e} particles/m^3".format(rho_solid/sigma_Ar**3))
print("T = {0:.3e} K".format(T_solid*eps_Ar/kB))
print("sigma = {0:3e} m ".format(sigma_Ar))
print("epsilon = {0:3e} J".format(eps_Ar))
print("time step = {0:3e} s".format(delta_Ar))
```

```python
#ENERGY
plt.errorbar(time_Ar,ekin_solid*eps_Ar,yerr=err_ekin_solid*eps_Ar, label='<K>')
plt.xlabel('Time [s]')
plt.ylabel('Energy [J]')
#plt.legend(loc='upper right')
plt.grid(True)
plt.title('Solid Argon: kinetic energy')
plt.show()

plt.errorbar(time_Ar,epot_solid*eps_Ar,yerr=err_epot_solid*eps_Ar, label='<V>')
plt.xlabel('Time [s]')
plt.ylabel('Energy [J]')
#plt.legend(loc='upper right')
plt.grid(True)
plt.title('Solid Argon: potential energy')
plt.show()

plt.errorbar(time_Ar,etot_solid*eps_Ar,yerr=err_etot_solid*eps_Ar, label='<E>')
plt.xlabel('Time [s]')
plt.ylabel('Energy [J]')
#plt.legend(loc='upper right')
plt.grid(True)
plt.title('Solid Argon: total energy')
plt.show()
```

```python
#TEMPERATURE
plt.errorbar(time_Ar,temp_solid*eps_Ar/kB,yerr=err_temp_solid*eps_Ar/kB, label='<T>')
plt.xlabel('Time [s]')
plt.ylabel('Temperature [K]')
#plt.legend()
plt.grid(True)
plt.title('Solid Argon: temperature')
plt.show()
```

```python
#PRESSURE
plt.errorbar(time_Ar,pres_solid*eps_Ar/sigma_Ar**3,yerr=err_pres_solid*eps_Ar/sigma_Ar**3, label='<p>')
plt.xlabel('Time [s]')
plt.ylabel('Pressure [Pa]')
#plt.legend()
plt.grid(True)
plt.title('Solid Argon: pressure')
plt.show()
```
