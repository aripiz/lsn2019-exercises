```python
import tensorflow as tf
from tensorflow import keras

from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D
from keras.optimizers import SGD, Adam, Adagrad

import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["figure.figsize"] = (9.5,6)
```

```python
seed=0
np.random.seed(seed) # fix random seed
tf.set_random_seed(seed)
```

## Importing MNIST dataset ##

```python
from keras.datasets import mnist

# input image dimensions
img_rows, img_cols = 28, 28 # number of pixels 

# output
num_classes = 10 # 10 digits

# the data, split between train and test sets
(X_train, Y_train), (X_test, Y_test) = mnist.load_data()
print('Training data: ',len(X_train))
print('Validation data: ',len(X_test))
```

```python
# cast floats to single precesion
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

# rescale data in interval [0,1]
X_train /= 255
X_test /= 255

# convert class vectors to binary class matrices, e.g. for use with categorical_crossentropy
Y_train = keras.utils.to_categorical(Y_train, num_classes)
Y_test = keras.utils.to_categorical(Y_test, num_classes)

# reshape data, depending on Keras backend
X_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
X_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1) 
print('X_train shape:', X_train[0].shape)
```

## Models construction and training##

The Convolutional Neural Network used is composed by the following layers:
- Convolutional 2D layer of 32 filters, kernel size 5x5 and _relu_ activation
- Max pooling 2D layer of size 2x2
- Convolutional 2D layer of 64 filters, kernel size 5x5 and _relu_ activation
- Max pooling 2D layer of pool size 2x2
- Dropout 0.25
- Dense layer of dimension 128 and _relu_ activation
- Dropout 0.5
- Output dense layer of dimension 10 and _softmax_ activation

```python
def create_CNN():
    # instantiate model
    model = Sequential()
    # add first convolutional layer with 32 filters (dimensionality of output space)
    model.add(Conv2D(32, kernel_size=(5, 5), activation='relu', input_shape=input_shape))

    model.add(MaxPooling2D(pool_size=[2, 2], strides=2))
    
    model.add(Conv2D(64, kernel_size=(5, 5), activation='relu'))
    
    model.add(MaxPooling2D(pool_size=(2, 2), strides=2)) 
    
    model.add(Dropout(0.25))
    model.add(Flatten())    
    model.add(Dense(128, activation='relu'))

    model.add(Dropout(0.5))
   
    # soft-max layer
    model.add(Dense(num_classes, activation='softmax'))
    
    # compile the model
    model.compile(loss=keras.losses.categorical_crossentropy,
                  optimizer='SGD',
                  metrics=['accuracy'])
    return model
```

```python
# training parameters
batch_size = 32
epochs = 5
```

```python
# create the deep neural net
model_SGD = create_CNN()
model_SGD.summary()
```

```python
# train DNN and store training info in history
history_SGD = model_SGD.fit(X_train, Y_train,
          batch_size=batch_size,
          epochs=epochs,
          verbose=1,
          validation_data=(X_test, Y_test))
```

```python
# evaluate model

# print performance
print('CNN score:')
score_SGD = model_SGD.evaluate(X_test, Y_test, verbose=1)
print('Test loss:', score_SGD[0])
print('Test accuracy:', score_SGD[1])

# look into training history

# summarize history for accuracy
plt.plot(history_SGD.history['acc'])
plt.plot(history_SGD.history['val_acc'])
plt.ylabel('Accuracy')
plt.title('CNN accuracy with SGD optmizer')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()

# summarize history for loss
plt.plot(history_SGD.history['loss'])
plt.plot(history_SGD.history['val_loss'])
plt.ylabel('Loss')
plt.title('CNN loss with SGD optmizer')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```
