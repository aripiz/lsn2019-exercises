# Exercise 06.1 #

Hamiltonian for a system of $N$ spins with nearest neighbours interaction (1D Ising Model):
$$
H = 
-J\sum_{i=1}^N s_i s_{i+1}
-\frac{h}{2}\sum_{i=1}^N s_i + s_{i+1} 
$$
where $\{ s_i \}_{i=1,N}$ are spin variables taking values in $\{-1;1\}$, $J>0$ (ferromagnetic model).

Periodic boundary conditition - such that $s_{N+1} = s_{1}$ - and $\mu_B=1$, $k_B=1$ units are assumed.

100 blocks of 1000 Metropolis steps at ($T=1.5, h=0$), starting from a randomly generated configuration and a ground state configuration.

## Measurments ## 

20 blocks of 10000 Metropolis steps.

Estimation as a function of $T$ with increments of $0.01$ in the range $\left[0.5;2.0\right]$ of:
1. Internal energy $U(N,T) = \langle H \rangle_{h=0}$
2. Heat capacity $C(N,T) =  \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. Magnetic susceptibility $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. Magnetization $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

All quantities are meant per spin ($N=50$), except the Magnetization. $J=1$.

Results from Gibbs and Metropolis sampling are compared with expected values for 1D Ising Model.


```python
#EXACT RESULTS FROM 1D ISING SOLUTION
points=150
T = np.linspace(0.5,1.99,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)

h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
```
